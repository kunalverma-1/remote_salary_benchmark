```python
df_h1b = pd.read_excel('../data/raw_data/LCA_Dislclosure_Data_FY2026_Q2.xlsx')
print(df_h1b.shape)
print(df_h1b.columns.tolist())
```

```python
cols_needed = ['CASE_STATUS', 
               'JOB_TITLE',
               'EMPLOYER_NAME',
               'EMPLOYER_CITY',
               'EMPLOYER_STATE',
               'EMPLOYER_COUNTRY',
               'WAGE_RATE_OF_PAY_FROM',
               'WAGE_UNIT_OF_PAY'
              ]
df_h1b = df_h1b[cols_needed]
print(df_h1b.shape)
df_h1b.head(10)
```

```python
df_ppp = pd.read_csv('../data/raw_data/API_PA.NUS.PPP_DS2_en_csv_v2_281504.csv',skiprows = 4)
print(df_ppp.shape)
df_ppp.head()
```

```python
df_ppp = df_ppp[['Country Name', 'Country Code', '2024']]
df_ppp.columns = ['country', 'country_code', 'ppp_2024']
print(df_ppp.shape)
df_ppp.head(10)
```

```python
df_levels = pd.read_csv('../data/raw_data/salaries.csv')
print(df_levels.shape)
print(df_levels.columns.tolist())
df_levels.head(10)
```

```python
cols_needed = ['job_title',
               'experience_level',
               'salary_in_usd',
               'employee_residence',
               'employment_type']
df_levels = df_levels[cols_needed]
print(df_levels.shape)
df_levels.head(10)
```

```python
df_h1b.to_csv('../data/raw_data/h1b_raw.csv', index=False)
df_ppp.to_csv('../data/raw_data/ppp_raw.csv', index=False)
df_levels.to_csv('../data/raw_data/levels_raw.csv', index=False)

print("all 3 datasets saved")
```
